```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# Data
data = {
  "DNN": ["MobileNetV1", "MobileNetV1", "MobileNetV1", "MobileNetV1", "MobileNetV1", "MobileNetV1",
      "MobileNetV2", "MobileNetV2", "MobileNetV2", "MobileNetV2", "MobileNetV2", "MobileNetV2",
      "InceptionV1", "InceptionV1", "InceptionV1", "InceptionV1", "InceptionV1", "InceptionV1",
      "Resnet18", "Resnet18", "Resnet18", "Resnet18", "Resnet18", "Resnet18"],
  "Hardware Setup": ["CPU 1T", "SA 1T", "VM 1T", "CPU 2T", "SA 2T", "VM 2T",
             "CPU 1T", "SA 1T", "VM 1T", "CPU 2T", "SA 2T", "VM 2T",
             "CPU 1T", "SA 1T", "VM 1T", "CPU 2T", "SA 2T", "VM 2T",
             "CPU 1T", "SA 1T", "VM 1T", "CPU 2T", "SA 2T", "VM 2T"],
  "Overall": [776, 231, 264, 402, 159, 178,
        702, 279, 332, 371, 191, 222,
        1533, 342, 380, 853, 342, 366,
        1894, 537, 687, 1051, 537, 682]
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the data for grouped bar chart
pivot_df = df.pivot(index="DNN", columns="Hardware Setup", values="Overall")

# Sort columns by thread (1T and 2T)
sorted_columns = sorted(pivot_df.columns, key=lambda col: (col.split()[1], col.split()[0]))
pivot_df = pivot_df[sorted_columns]

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.15
x = np.arange(len(pivot_df.index))  # the label locations

for i, column in enumerate(pivot_df.columns):
  ax.bar(x + i * bar_width, pivot_df[column], bar_width, label=column)

# Add labels, title, and legend
ax.set_xlabel("DNN Models")
ax.set_ylabel("Overall (ms)")
ax.set_title("Performance Comparison by DNN Model and Hardware Setup")
ax.set_xticks(x + bar_width * (len(pivot_df.columns) - 1) / 2)
ax.set_xticklabels(pivot_df.index)
ax.legend(title="Hardware Setup")

plt.tight_layout()
plt.show()
```
