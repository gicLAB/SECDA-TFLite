```python
# load two csv files and compare them
import csv
import os

def load_csv(filename):
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        l = list(reader)
        l = l[1:] if len(l[0]) == 0 else l
        return l
  
def compare_csv(csv1, csv2):
    differ = []
    if len(csv1) != len(csv2):
        print("number of rows differ")
        return False

    if len(csv1[0]) != len(csv2[0]):
        print("number of columns differ")
        return False
    
    for row in range(len(csv1)):
        for col in range(len(csv1[0])):
            if csv1[row][col] != csv2[row][col]:
                differ.append((row, col))
    return differ

def visualise_csv(csv, differ):
    # print the csv with the differences highlighted
    for row in range(len(csv)-1):
        for col in range(len(csv[0])-1):
            number = csv[row][col]
            str_number = str(number)
            if "-" not in str_number:
                str_number = "+" + str_number
            if (row, col) in differ:
                print(f"\033[91m{str_number:4}\033[0m", end=" ")
            else:
                print(f"{str_number:4}", end=" ")
        print()

def get_csv_files(folder):
    csv_files = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(".csv"):
                csv_files.append(os.path.join(root, file))
    
    # create a list of pairs of csv files if they are start with the same name before second underscore
    csv_files = sorted(csv_files)
    csv_files = [csv_files[i:i+2] for i in range(0, len(csv_files), 2)]
    return csv_files

# create custom sort function
def sort_files(file):
    return int(file[0].replace(fol, "").split("_")[0])

pynq = "/run/user/1000/gvfs/smb-share:server=192.168.0.24,share=xilinx/Workspace/secda_benchmark_suite/aData/"
# fol = "/home/jude/Workspace/SECDA-TFLite_v1.2/tensorflow/aData/fc/" 
# fol = "/home/jude/Workspace/SECDA-TFLite_v1.2/tensorflow/aData/conv/" 
fol = pynq + "mm2im/"
fol = "/run/user/1000/gvfs/smb-share:server=192.168.0.24,share=xilinx/Workspace/secda_benchmark_suite/aData/mm2im/"

# pynq folder needs to be samba active be used
files = (get_csv_files(fol))
files = sorted(files, key=sort_files)


for file in files:
    print("comparing: " + file[0] + " and " + file[1])
    csv1 = load_csv(file[0])
    csv2 = load_csv(file[1])
    differ = compare_csv(csv1, csv2)
    if differ == False:
        print("matrix shape differs")
    else:
        if len(differ) == 0:
            # print("layer:" +  file[0].replace(fol, "").split("_")[0] + ": " + str(len(differ)))
            continue
        # diff_filp = [(x[1],x[0], csv1[x[0]][x[1]], csv2[x[0]][x[1]]) for x in differ]
        diff_filp = [(x[1],x[0], int(csv1[x[0]][x[1]])  - int(csv2[x[0]][x[1]])) for x in differ]
        print("layer:" +  file[0].replace(fol, "").split("_")[0] + ": " + str(len(differ)))
        for  idx,x in enumerate(diff_filp):
            print(x)
            if idx > 10:
                break
        # visualise_csv(csv1, differ)
        break
    
  
```
